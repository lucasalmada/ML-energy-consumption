```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
# Etapa 1: Ler o arquivo "data.csv"
df = pd.read_csv("data.csv")

# Etapa 2: Excluir a coluna 'id' e valores NAN
df.drop(["id",'Unnamed: 32'], axis=1, inplace=True)
df = df.dropna()

# Etapa 3: Converter a coluna 'diagnosis' em categórica, M = 1, B = 0
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Definir a variável target e as features
target = 'diagnosis'
features = df.columns[df.columns != target]

# Etapa 4: Dividir em treino e teste (70% treino, 30% teste)
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.3, random_state=42)

# Etapa 5: Rodar um algoritmo de classificação (Random Forest neste exemplo)
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Etapa 6: Avaliar o modelo
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Acurácia:", accuracy)
print("Precisão:", precision)
print("Recall:", recall)
```
